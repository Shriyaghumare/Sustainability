```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
# sample data of 5 countries of 5 years
data = {
    'Country':['Country 1','Country 2','Country 3', 'Country 4','Country 5'],
    'GDP':[45000,50000,39000,10000,20000],
    'Unemployment_rate':[5.7,9.0,4.6,6.0,12.9],
    'Education_rate': [0.89,0.75,0.68,0.83,0.74],
    'Health_rate': [0.88,0.56,0.78,0.80,0.90]
}
df = pd.DataFrame(data)
df  
```

```python
averages = df[['GDP','Unemployment_rate','Education_rate','Health_rate']].mean()
print(f"Average Indicators: \n{averages}")
```

```python
plt.figure(figsize=(6,5))
plt.bar(df['Country'],df['GDP'], color='green')
plt.title('GDP per Capital of Countries')
plt.xlabel('Country')
plt.ylabel('GDP per capital')
plt.grid(True)
plt.show()
```

```python
#Education v/s Health rate of country
df.plot(x='Country', y=['Education_rate','Health_rate'], kind ='bar', figsize=(6,5), colormap='coolwarm')
plt.title ('Education v/s Health rates of Countries')
plt.ylabel('Rate (0 to 1)')
plt.grid(True)
plt.show()
```

```python
#Unemployment rate
plt.plot(df['Country'], df['Unemployment_rate'], marker='o', linestyle='-', color='pink')
plt.title('Unemployment Rate of Countries')
plt.ylabel('Unemployment Rate (in %)')
plt.grid(True)
plt.show()
```

```python
print("Country with Highest Education rate :")
print(df.loc[ df['Education_rate'].idxmax(), 'Country'])
```

```python
print("Country with Lowest Education rate :")
print(df.loc[ df['Education_rate'].idxmin(), 'Country'])
```
